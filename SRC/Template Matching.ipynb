```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
import os, shutil
```

```python
def plot(img, cvtColor=True):
    plt.figure(figsize=(10, 10))
    if cvtColor:
        plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
        return
    plt.imshow(img)
    return
```

```python
image_lis = os.listdir('images/')
image_lis = sorted(image_lis, key = lambda x:int(x.split('.')[0]))
```

```python
img = cv2.imread(f'images/{image_lis[0]}')

# Try matching this particular portion
# plot(img[250:400, 150:300,:])

img1 = img[300:400, 170:270,:]
plot(img1)
w, h, _ = img1.shape
```

```python
img2 = cv2.imread(f'images/{image_lis[1]}')
plot(img2)
```

```python
res = cv2.matchTemplate(img2, img1, cv2.TM_CCOEFF_NORMED)
min_val, max_val, min_loc, max_loc = cv2.minMaxLoc(res)
```

```python
top_left = max_loc
bottom_right = (top_left[0] + w, top_left[1] + h)
```

```python
plot(cv2.rectangle(img2,top_left, bottom_right, 255, 2))
```
